```python
import numpy as np
import matplotlib.pyplot as plt

def convex_hull(points):
    n = len(points)
    if n < 3:
        return []

    hull = []
    l = np.argmin(points[:, 0])

    while True:
        hull.append(points[l])
        q = (l + 1) % n

        for r in range(n):
            if orientation(points[l], points[q], points[r]) < 0:
                q = r

        l = q

        if l == np.argmin(points[:, 0]):
            break

    return np.array(hull)

def orientation(p, q, r):
    return (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])

belanda_points = np.array([[1, 2], [3, 5], [5, 8]])
indonesia_points = np.array([[1, 2], [4, 5], [3, 8]])

belanda_hull = convex_hull(belanda_points)
indonesia_hull = convex_hull(indonesia_points)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(belanda_points[:, 0], belanda_points[:, 1], color='blue', label='Bahasa Belanda')
plt.plot(belanda_hull[:, 0], belanda_hull[:, 1], color='red', label='Convex Hull')
plt.title('Bahasa Belanda')
plt.legend()

plt.subplot(1, 2, 1)
plt.scatter(indonesia_points[:, 0], indonesia_points[:, 1], color='green', label='Bahasa Indonesia')
plt.plot(indonesia_hull[:, 0], indonesia_hull[:, 1], color='red', label='Convex Hull')
plt.title('Bahasa Indonesia')
plt.legend()

plt.tight_layout()
plt.show()
```
