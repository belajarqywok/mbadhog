```python
import os
import re
import time
import requests
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
```

```python
# Browser Options
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-gpu')
options.add_argument("window-size=1200x600")
options.add_argument('--disable-web-security')
options.add_argument('--allow-running-insecure-content')
options.add_argument('--allow-cross-origin-auth-prompt')

browser = webdriver.Chrome(options=options)

payload = {
    "kelurahan": "Pinang Ranti",
    "kecamatan": "Makasar",
    "city": "Jakarta Timur",
    "province": "DKI Jakarta"
}

# Get Name Location --
def get_name_location(browser, iter):
  try:
    name_location_xpath = f'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]/div[{iter + 1}]/div/div[2]/div[4]/div[1]/div/div/div[2]/div[1]/div[2]'
    name_location = browser.find_element('xpath', name_location_xpath)
    return name_location.text
  except NoSuchElementException:
    return None

# Get Rating Location --
def get_rating_location(browser, iter):
  try:
    rating_location_xpath = f'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]/div[{iter + 1}]/div/div[2]/div[4]/div[1]/div/div/div[2]/div[3]/div/span[2]/span/span[1]'
    rating_location = browser.find_element('xpath', rating_location_xpath)
    return re.sub(r'(\d+),(\d+)', r'\1.\2', rating_location.text)
  except NoSuchElementException:
    return None

# Get Address Location --
def get_address_location(browser, iter):
  try:
    address_location_xpath = f'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]/div[{iter + 1}]/div/div[2]/div[4]/div[1]/div/div/div[2]/div[4]/div[1]/span[2]/span[2]'
    address_location = browser.find_element('xpath', address_location_xpath)
    return address_location.text
  except NoSuchElementException:
    return None

# Get Contact Location (Owner) --
def get_contact_location(browser, iter):
  try:
    contact_location_xpath = f'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]/div[{iter + 1}]/div/div[2]/div[4]/div[1]/div/div/div[2]/div[4]/div[2]/span[2]/span[2]'
    contact_location = browser.find_element('xpath', contact_location_xpath)
    return contact_location.text
  except NoSuchElementException:
    return None

# Get Geolocation
def get_geolocation(url):
  return {
      'lat': re.search(r'!3d(-?\d+\.\d+)', url).group(1),
      'lon': re.search(r'!4d(-?\d+\.\d+)', url).group(1)
  }

def scrape(payload: dict) -> dict:
  search = (
    "warteg terdekat, terbaik, dan terenak di " +
    f"{payload.get('kelurahan')} {payload.get('kecamatan')} " +
    f"{payload.get('city')} {payload.get('province')}"
  )

  url = f"https://www.google.com/maps/search/{search}"
  browser.get(url)

  for iter in range(100):
      try:
        location_object_xpath = f'//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]/div[{iter + 1}]/div/a'
        location_object = browser.find_element('xpath', location_object_xpath)
        location_object.click()

        # url
        url = location_object.get_attribute('href')

        # name
        name = get_name_location(browser, iter)

        # rating
        rating = get_rating_location(browser, iter)

        # address
        address = get_address_location(browser, iter)

        # contact
        contact = get_contact_location(browser, iter)

        # geolocation
        geolocation = get_geolocation(url)
        lat = geolocation.get('lat')
        lon = geolocation.get('lon')

        print(f"name: {name}")
        print(f"rating: {rating}")
        print(f"address: {address}")
        print(f"contact: {contact}")
        print(f"url: {re.sub(r'hl=[A-Za-z-]+', 'hl=id', url)}")
        print("geolocation :")
        print(f"  - lat: {lat}")
        print(f"  - lon: {lon}")
        print(f"  - geo_url: https://www.google.com/maps/search/{lat},{lon}/")
        print("\n")


        if not bool((iter + 1) % 5):
          scroll_xpath = '/html/body/div[3]/div[8]/div[9]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]'
          browser.execute_script(
              "var el = document.evaluate(" +
                f"'{scroll_xpath}', " +
                "document, null, XPathResult.FIRST_ORDERED_NODE_TYPE, null " +
              ").singleNodeValue;" +
              "el.scrollBy(0, 5000);"
          )

      except NoSuchElementException:
        time.sleep(1)
      finally:
        continue

scrape(payload)
```
